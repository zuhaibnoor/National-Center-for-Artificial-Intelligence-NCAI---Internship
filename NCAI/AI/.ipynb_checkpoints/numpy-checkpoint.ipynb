# NUMPY
Numpy is a python library used for manipulating large arrays that are multi-dimensional. It also the basis of pandas.

```python
import numpy as np
```

## Numpy array
A <b>python list</b> is a container that can hold data of different types. It can have variable length. A <b>Numpy array</b> on the other hand has a fixed size and contains the items of same data type. The data can be accessed like in a python list through indexes. 

```python
#Creating a numpy array
np_array = np.array([1,2,3,4])

#to obtain the type of array
print(type(np_array))

#to obtain the type of elements in the array we use dtype attribute
print(np_array.dtype) #will return numpy.ndarray
```

## Attributes of a Numpy array
Some common attributes of a numpy array are:
<ul>
    <li><b>size:</b> The size of an array or the number of elements in an array.</li>
    <li><b>dtype:</b> The datatype of the elements in the array.</li>
    <li><b>ndim:</b> Determines the dimension of an array. In this case 1.</li>
    <li><b>shape:</b> It is a tuple of integers that determins the size of array in each dimension.</li>    
</ul>

## Vectors
A vector is an array with only one dimension. For eg: [0,1],[1,0]
### Vector addition and subtraction



```python
#vector addition can be done with only one line of code
x = np.array([1,0])
y = np.array([0,1])
z = x+y
print(z)

#similarly subtraction can be done
z = z-x
print(z)
```

### Vector multiplication




```python
x = np.array([1,2])
z = 2*x
print(z)
```

```python
# in python list, such operation will be slower
myls = [1,2]
for i in range(len(myls)):
    myls[i] = 2*(i+1)
print(myls)
```

### Vector Product



```python
u = np.array([1,2])
v = np.array([3,2])
z = u*v
print(z)
```

### Dot product

we can use the <b>np.dot(u,v)</b> for this purpose.

```python
u = np.array([1,2])
v = np.array([3,1])
print(np.dot(u,v))
```

## Adding constant to a numpy array
To add a constant to a numpy array we can simply define it and add +1.

```python
u = np.array([1,2,3,-1])
print(u+1)
```

## Other functions for numpy arrays
<ul>
    <li><b>.mean():</b> We can apply this fucntion to fund out the mean value in a numpy array.</li>
    <li><b>.max(): </b>To find out the largest value in a numpy array.</li>
</ul>

```python
u = np.array([2,4])
print(u.mean())
print(u.max())
```

<li><b>np.pi :</b> To access the value of pi</li>

```python
arr = np.array([0, np.pi/2, np.pi])

#to find out the sine of each element in the numpy array
print(np.sin(arr))
```

<li><b>.linespace(starting_point, ending_point, no_of_digits_to_generate) :</b> It is used to generate a numpy array by specifying the starting point and ending point. The no_of_digits_to_generate specifies the no of digits to generate with equal intervals.</li>

```python
print(np.linspace(-2,2, num = 5))
```

```python
print(np.linspace(0,10, num = 11))
```

```python
# to generate a numpy array containing 100 values from pi to 2 pi
x = np.linspace(0,(np.pi)*2, num = 100)
print(x)
```

```python
y = np.sin(x)
print(y)
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
```

## Two dimensional numpy array
To make a 2d numpy array, we make a 2d python list and then cast it into the numpy's array function.

```python
arr_2d = [[1,2,3],[4,5,6],[7,8,9]]
arr_2d = np.array(arr_2d)
arr_2d
```

```python
# to obtain the dimension of this list
arr_2d.ndim
```

```python
# the shape attribute can be used to find out the number of nested lists and the dimensions in a tuple
print(arr_2d.shape)
#for total number of elements
print(arr_2d.size)
```

### Accessing the elements in a 2d numpy array
we can access the elements in a numpy array through double square brackets or a single one too. For eg: <b>a[0][1]</b> or <b>a[0,1]</b> will access the same element from a numpy array. The second method won't work in a simple python 2d list.

```python
arr_2d[0,1]
```

#### we can also apply <b>slicing</b> on a 2d numpy array

```python
print(arr_2d[0,0:2])
#or
print(arr_2d[0][0:2])
```

```python
arr_2d[0:2,2]
```

## Addition of Matrices or 2d numpy array
We define two 2d arrays or matrices then we can simply use the addition operator to get the sum of all the respective elements


```python
x = np.array([[1,2],[3,4]])
y = np.array([[5,6],[7,8]])
x+y
```

## Multiplication of a matrix by a number
A matric can be defined and then simply multiplied by a number by using the <b>*</b> operator.

```python
x = np.array([[1,2],[3,4]])
x*2
```

## Multiplying the corresponding elements of a 2d numpy array
Define the 2 matrices and use the <b>*</b> operator to multiplying the corresponding elements of the matrices.



```python
x = np.array([[1,0],[0,1]])
y = np.array([[2,1],[1,2]])
z = x*y
z
```

## Multiplying two matrices using dot method and @ operator


```python
z = np.dot(x,y)
z
```

```python
z = x @ y
z
```
